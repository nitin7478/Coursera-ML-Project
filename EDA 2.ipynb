## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
stmt = ibm_db.exec_immediate(conn, "select DISTINCT LAUNCH_SITE from SPACEX")
result = ibm_db.fetch_both(stmt)
print(result)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
stmt = ibm_db.exec_immediate(conn, "SELECT * FROM spacex WHERE launch_site LIKE 'CCA%' LIMIT 5;")
result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
stmt = ibm_db.exec_immediate(conn, "SELECT SUM(PAYLOAD_MASS_KG_) AS TOTAL_PAYLOAD  FROM spacex WHERE customer= 'NASA (CRS)';")
result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
stmt = ibm_db.exec_immediate(conn, "SELECT AVG(PAYLOAD_MASS_KG_) AS TOTAL_PAYLOAD  FROM spacex WHERE BOOSTER_VERSION = 'F9 v1.1';")
result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
stmt = ibm_db.exec_immediate(conn, "SELECT MIN(DATE) AS FIRST_SUCCESSFULL_LANDING FROM SPACEX WHERE LANDING_OUTCOME = 'Success (ground pad)';")
result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
stmt = ibm_db.exec_immediate(conn, "SELECT  BOOSTER_VERSION FROM SPACEX WHERE LANDING_OUTCOME = 'Success (drone ship)' AND PAYLOAD_MASS_KG_ > 4000 AND PAYLOAD_MASS_KG_ <6000;")
result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
stmt = ibm_db.exec_immediate(conn, "SELECT COUNT(MISSION_OUTCOME) AS SUCCESSFUL_MISSION_OUTCOME FROM spacex where  MISSION_OUTCOME LIKE 'Success%'")
result = ibm_db.fetch_both(stmt)
print(result)
```

```python
stmt = ibm_db.exec_immediate(conn, "SELECT COUNT(MISSION_OUTCOME) AS FAILURE_MISSION_OUTCOME FROM spacex where  MISSION_OUTCOME LIKE 'Failure%'")
result = ibm_db.fetch_both(stmt)
print(result)
```

```python

stmt = ibm_db.exec_immediate(conn, "SELECT COUNT(MISSION_OUTCOME) AS TOTAL_SUCCESSFULL_AND_FAILURE_MISSION_OUTCOME FROM spacex WHERE MISSION_OUTCOME LIKE 'Failure%' OR MISSION_OUTCOME LIKE 'Success%';")

result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python

stmt = ibm_db.exec_immediate(conn, "SELECT DISTINCT BOOSTER_VERSION AS BOOSTER_VERSIONS FROM SPACEX WHERE PAYLOAD_MASS_KG_ =(SELECT MAX(PAYLOAD_MASS_KG_) FROM SPACEX);")

result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 9


##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python

stmt = ibm_db.exec_immediate(conn, " SELECT BOOSTER_VERSION, LAUNCH_SITE FROM SPACEX WHERE DATE LIKE '2015-%' AND LANDING_OUTCOME = 'Failure (drone ship)';")

result = ibm_db.fetch_both(stmt)
print(result)
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python

stmt = ibm_db.exec_immediate(conn, "SELECT LANDING_OUTCOME AS LandingOutcome, COUNT(LANDING_OUTCOME) AS TotalCount FROM SPACEX WHERE DATE BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY LANDING_OUTCOME ORDER BY COUNT(LANDING_OUTCOME) DESC")

result = ibm_db.fetch_both(stmt)
print(result)
```
